```python
generation.index.get_level_values(0).unique()
```

```python
generation.index.unique()
```

```python
pct_to_drop = [col for col in generation.columns if 'pct' in col]
gen_twh = generation.copy()
gen_twh.drop(columns=pct_to_drop, inplace=True)
```

```python
gen_sum = gen_twh.groupby('year').sum()
gen_sum
```

```python
ax = gen_sum.plot.area(stacked=True)
fig = ax.get_figure()
fig.set_size_inches(10, 6)
plt.legend(loc='upper right')
plt.show()
```
